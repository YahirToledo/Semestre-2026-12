## Introducción

En el desarrollo de la robótica moderna, es fundamental contar con un dominio básico de herramientas digitales que permiten documentar, programar y compartir proyectos.

- <img src="imagenes/Ubuntu-logo.png" height=60/> **Ubuntu** es un sistema operativo basado en Linux, muy usado en investigación y robótica gracias a su estabilidad y compatibilidad con librerías como ROS. El manejo de la terminal es esencial para automatizar tareas y ejecutar programas de manera eficiente.



- <img src="imagenes/Jupyter-logo.png" height=60/> **Jupyter** es una herramienta interactiva que facilita combinar texto, código y resultados en un mismo documento. Se utiliza ampliamente en ciencia de datos, aprendizaje automático y simulación robótica. Para texto utiliza el formato **Markdown**.



- <img src="imagenes/Python-logo.png" height=60/> **Python** es uno de los intérpretes más usados en robótica por su simplicidad y su gran ecosistema de librerías. En Jupyter, se puede combinar código Python con explicaciones en texto.



- <img src="imagenes/Git-logo.png" height=60/> **Git** es un sistema de control de versiones que permite llevar un historial de cambios en los archivos de una carpeta (repositorio), así como recuperar versiones anteriores, y trabajar en paralelo en "*ramas*" sin perder información. 



- <img src="imagenes/Github-logo.png" height=60/> **Github** es una plataforma en línea que utiliza git para almacenar repositorios. Cuenta con herramientas para colaboración en equipo, documentación y despliegue.



## Cuestionario previo

### Responder de forma breve las siguientes preguntas:

- En linux, ¿a qué se refieren los términos sh, bash y shell?
> En Linux, el shell es el programa que interpreta los comandos y sirve como interfaz entre el usuario y el sistema operativo. Dentro de este contexto, sh (Bourne Shell) fue uno de los primeros intérpretes estándar en Unix, muy básico pero funcional, mientras que bash (Bourne Again Shell) es su evolución más completa y actualmente el shell más usado, ya que incluye mejoras como historial, autocompletado y estructuras de control avanzadas.
- ¿Python se considera un lenguaje de programación? ¿Por qué?
> Python sí se considera un lenguaje de programación, ya que cuenta con reglas de sintaxis y semántica, permite escribir algoritmos, manipular datos y desarrollar programas completos. Además, es multiparadigma y muy versátil, utilizado en áreas como desarrollo web, inteligencia artificial, ciencia de datos y automatización.
- ¿Qué lenguajes o intérpretes se pueden usar en Jupyter?
> Aunque Jupyter nació para Python, actualmente soporta varios lenguajes mediante kernels.
Entre los más comunes están: Python, R, Julia, Octave/Matlab, C/C++ y Java, aunque se pueden instalar muchos más según las necesidades.
- ¿Cuál es la diferencia entre Git y GitHub?
> Git es una herramienta de control de versiones que funciona en la computadora y permite manejar el historial del código y trabajar en ramas.
GitHub es una plataforma en la nube que usa Git como base para colaborar, compartir proyectos y almacenar repositorios en línea.

### 3. Python

Crear en python las siguientes celdas:

- Definir una clase "Calculadora" con métodos "suma" y "multiplicación" que tomen dos parámetros y regresen el resultado de la operación
- Definir una clase "CalculadoraMultiple" con métodos "suma" y "multiplicación" que tomen cualquier cantidad de parámetros y regresen el resultado de la operación 
- Instanciar ambas clases en una nueva celda y comprobar su funcionamiento

~~~python
#!/user/bin/env python3

""" Calculadora en Python """

class Calculadora:
    """Clase para operaciones con dos números."""
    def suma(self, a, b):
        return a + b

    def multiplicacion(self, a, b):
        return a * b


class CalculadoraMultiple:
    """Clase para operaciones con múltiples números."""
    def sumaMultiple(self, *numeros):
        return sum(numeros)

    def multiplicacionMultiple(self, *numeros):
        resultado = 1
        for n in numeros:
            resultado *= n
        return resultado


def menu():
    calc = Calculadora()
    calcMulti = CalculadoraMultiple()

    while True:
        print("\n===== MENÚ DE CALCULADORA =====")
        print("1. Suma (2 números)")
        print("2. Multiplicación (2 números)")
        print("3. Suma múltiple")
        print("4. Multiplicación múltiple")
        print("5. Salir")

        opcion = input("\nSelecciona una opción: ").strip()

        if opcion == "1":
            a = float(input("Ingresa el primer número: "))
            b = float(input("Ingresa el segundo número: "))
            print(f"\nResultado de la suma: {calc.suma(a, b)}")

        elif opcion == "2":
            a = float(input("Ingresa el primer número: "))
            b = float(input("Ingresa el segundo número: "))
            print(f"\nResultado de la multiplicación: {calc.multiplicacion(a, b)}")

        elif opcion == "3":
            numeros = [float(x) for x in input("Ingresa los números separados por espacio: ").split()]
            print(f"\nResultado de la suma múltiple: {calcMulti.sumaMultiple(*numeros)}")

        elif opcion == "4":
            numeros = [float(x) for x in input("Ingresa los números separados por espacio: ").split()]
            print(f"\nResultado de la multiplicación múltiple: {calcMulti.multiplicacionMultiple(*numeros)}")

        elif opcion == "5":
            print("\nSaliendo de la calculadora. ¡Hasta luego!")
            break

        else:
            print("\n Opción no válida. Intenta de nuevo.")
    menu()
~~~

## Análisis de resultados

¿Qué utilidad tiene el uso de un repositorio en línea para código?
> Un repositorio en línea permite almacenar y gestionar el código en la nube, lo que facilita el trabajo colaborativo entre varios desarrolladores. Además, ofrece control de versiones, seguridad de respaldo, acceso desde cualquier lugar y herramientas para revisión de cambios e integración continua.

¿Cuál es la convención de ángulos que utilizan los archivos URDF para los ángulos de las juntas respecto al eslabón padre?
> Los archivos URDF utilizan la convención de ángulos en radianes, siguiendo la regla de la mano derecha para definir el giro de las juntas respecto al eje local del eslabón padre.

¿Qué utilidad tiene describir un robot en un archivo URDF?
> El URDF permite definir la estructura completa de un robot (eslabones, juntas, geometría, masas, colisiones y visualización). Esto es útil para simular en entornos como Gazebo o RViz, facilitar el control cinemático y dinámico, y estandarizar la comunicación con diferentes frameworks de robótica como ROS.
